```python
d={'anu':50,'bithu':40,'manu':35,'daya':20,'meenu':34,'serah':45}
x=d.keys()
y=d.values()
plt.bar(x,y,color="green",width=0.3)
plt.xlabel("student name")
plt.ylabel("total mark")
plt.title("students detail")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,5*np.pi,1)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,z)
plt.xlabel("values")
plt.ylabel("sin and cos")
plt.title("plot sin and cos graph")
plt.legend(['sin(x)','cos(x)'],loc='upper right')
plt.show()
```
